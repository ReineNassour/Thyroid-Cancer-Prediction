### DATASET DICTIONARY

**Age:** The age of the person in years.

**Gender:** The sex of the person, either Male (M) or Female (F).

**Smoking Hx:** A history of smoking; "Yes" means the person has smoked in the past, and "No" means they haven't.

**Smoking Hx Radiotherapy:** Whether the person has received radiotherapy treatment in the past and also has a history of smoking; "Yes" means both, and "No" means neither or just one.

**Thyroid Function:** The current condition of the thyroid gland; can be normal (Euthyroid) or show signs of disorders like hyperthyroidism.

**Physical Examination:** The results of a physical examination related to the thyroid, often describing the type of goiter (lumps or swelling in the thyroid).

**Adenopathy:** The presence of enlarged lymph nodes near the thyroid; "Right" or "Left" refers to the side of the body, and "None" means no enlargement.

**Pathology:** The type of thyroid disease found, such as different types of thyroid cancer (e.g., Micropapillary, Papillary, Hurthel cell).

**Focality:** Describes whether the thyroid disease affects one area (Uni-Focal) or multiple areas (Multi-Focal) of the thyroid.

**Risk:** The level of risk for the person's thyroid condition, usually categorized as "Low" or "High."

**T:** The stage of the thyroid tumor in terms of its size and extent. For example, T1a means a small tumor, while T4b indicates an advanced stage.

**N:** Describes whether nearby lymph nodes are affected by cancer; N0 means no affected lymph nodes, and N1b means some are affected.

**M:** Indicates whether cancer has spread to other parts of the body; M0 means no distant spread, while M1 means it has spread.

**Stage:** The overall stage of the cancer, combining T, N, and M, such as Stage I, Stage IVA, or IVB, which helps determine the severity.

**Response:** The person's response to treatment. It can be "Excellent" (good outcome), "Indeterminate" (uncertain outcome), or "Incomplete" (not fully treated or controlled).

**Recurred:** Whether the thyroid condition or cancer has come back after treatment; "Yes" means it has recurred, and "No" means it hasn't.

# Dividing Into training Set and Test Set

```python
x=df.drop(columns=['Recurred'])
y=df['Recurred']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# **1st Algorithm:** Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB()
model.fit(x_train,y_train)
y_pred_nb=model.predict(x_test)

from sklearn.metrics import accuracy_score,confusion_matrix,f1_score
print("Accuracy Score :",accuracy_score(y_test,y_pred_nb))
print("Confusion Matrix :",confusion_matrix(y_test,y_pred_nb))
print("F1 Score :",f1_score(y_test,y_pred_nb))

import time
st=time.time()
et=time.time()
#get the execution time
elapsed_time=et-st
print("Execution Time :",elapsed_time,'seconds')
```

# Plot the Target

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
y.value_counts().plot.pie()
plt.title("Pie Chart of Recurred")
```

# **2nd algorithm:**KNN

```python
from sklearn.neighbors import KNeighborsClassifier

maxk=min(15,len(x_train))
best_acc=0
best_k=0
k_values=[]
accuracy_values = []
for k in range(1,maxk+1):
  model=KNeighborsClassifier(n_neighbors=k)
  model.fit(x_train,y_train)
  y_pred=model.predict(x_test)
  acc=accuracy_score(y_test,y_pred)
  accuracy_values.append(acc)
  k_values.append(k)
  if acc>best_acc:
    best_acc=acc
    best_k=k
```

```python
knn=KNeighborsClassifier(n_neighbors=best_k)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)

from sklearn.metrics import accuracy_score,confusion_matrix,f1_score
print("Accuracy Score :",accuracy_score(y_test,y_pred_knn))
print("Confusion Matrix :",confusion_matrix(y_test,y_pred_knn))
print("F1 Score :",f1_score(y_test,y_pred_knn))

import time
st=time.time()
et=time.time()
#get the execution time
elapsed_time=et-st
print("Execution Time :",elapsed_time,'seconds')
```

```python
plt.plot(k_values ,accuracy_values)
plt.xlabel("K-value")
plt.ylabel("Accuracy")
plt.show()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)

y_pred_dt = dt.predict(x_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)

y_pred_rf = rf.predict(x_test)

print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf')
svm.fit(x_train, y_train)

y_pred_svm = svm.predict(x_test)

print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```
